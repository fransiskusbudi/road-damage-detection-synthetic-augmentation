```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Replace these with your actual image filenames
ground_truth_images = ["1.jpeg", "3.jpeg"]
prediction_images = ["2.jpeg", "4.jpeg"]

# Create 2x2 subplot grid (2 rows: samples, 2 columns: GT and prediction)
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Titles only on top row
axs[0, 0].set_title("Ground Truth", fontsize=16)
axs[0, 1].set_title("Prediction", fontsize=16)

for i in range(2):
    # Load and show images
    gt_img = mpimg.imread(ground_truth_images[i])
    pred_img = mpimg.imread(prediction_images[i])

    axs[i, 0].imshow(gt_img)
    axs[i, 1].imshow(pred_img)

    # Turn off axes for both
    axs[i, 0].axis('off')
    axs[i, 1].axis('off')

plt.tight_layout()
plt.savefig("comparison_grid.png")
plt.show()

```
